```python
import os
import pandas as pd

os.chdir("scRNA-seq")
```

```python
deseq_results = pd.read_csv("input_data/comparison_CalTRAP_mCN_2.csv")

bc_pval = .05 / deseq_results.shape[0]

deseq_filtered = deseq_results.loc[(deseq_results['pvalue'] < bc_pval) & (deseq_results['log2FoldChange'] > 2)]
deseq_filtered = deseq_filtered.sort_values('padj', ascending=True)
deseq_filtered.to_csv(f"data/enriched_genes_bc_pval_cutoff_and_lfc>2.csv", index=False)
```
