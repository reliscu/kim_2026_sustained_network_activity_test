```python
import pandas as pd
```

```python
# subset rMATS output to events of interest for downstream plotting

condition = "ctrl_vs_AR" # "ctrl_vs_AR", "ctrl_vs_naive", or "naive_vs_AR"
events_of_interest = pd.read_csv("input_data/targets_of_interest_v2.csv") # curated events to plot (must all be same splicing event type, e.g. "SE")
rmats_results = pd.read_csv(f"rmats_results/rmats/batch2/{condition}/SE.MATS.JC.txt", sep="\t")
```

```python
cols = ['exonStart_0base', 'exonEnd', 'upstreamES', 'upstreamEE', 'downstreamES', 'downstreamEE']
events_of_interest['Identifier'] = (events_of_interest[cols].astype(str).agg('-'.join, axis=1))
rmats_results['Identifier'] = (rmats_results[cols].astype(str).agg('-'.join, axis=1))
rmats_results_subset = rmats_results[rmats_results['Identifier'].isin(events_of_interest['Identifier'])]
rmats_results_subset.to_csv(f"data/rmats_events_of_interest_{condition}.txt", sep="\t", index=False)
```
